# Loading Datasets.

```python
import pandas as pd
import seaborn as sns
```

```python
tip = sns.load_dataset('tips')
flight = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
df= pd.read_csv('/content/tested.csv')
```

# Working with BI-Variate Analysis.

```python
# 1- Scatterplot() (Numerical - Numerical)

sns.scatterplot(data= tip, x= 'total_bill', y= 'tip')
```

```python
# In-order to get the no of Male, Female we do soo. (Multi-variate Analysis)

sns.scatterplot(data= tip, x= 'total_bill', y= 'tip', hue= 'sex', style= 'smoker')
```

```python
# 2- Barplot() (Numeical - Categorical)

sns.barplot(data= df, x= 'Pclass', y= 'Fare', hue= 'Sex')
```

```python
# 3- Boxplot() (Numerical - Categorical)

sns.boxplot(data= df, x= 'Sex', y= 'Age', hue= 'Survived')
```

```python
# 4- Distplot() (Numerical - Categorical)
import warnings
warnings.filterwarnings('ignore')

sns.distplot(df[df['Survived'] == 0]['Age'], hist= False) # This selects 'Age' col from filtered data & put 'Survived' equals to 0.
sns.distplot(df[df['Survived'] == 1]['Age'], hist= False)
```

```python
# 5- heatmap() (Categorical - Categorical)

sns.heatmap(pd.crosstab(df['Pclass'], df['Survived']))
```

```python
# 6- Clustermap() (Categorical - Categorical)

sns.clustermap(pd.crosstab(df['Pclass'], df['Survived']))
```

```python
# 7- Pairplot()

sns.pairplot(flight)
```
